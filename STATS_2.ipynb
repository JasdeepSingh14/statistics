***`Question 1: What is hypothesis testing in statistics?`***

***`Answer-`***

Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It helps determine whether there is enough evidence to support or reject a specific claim or assumption.

***`Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?`***

***`Answer-`***

The null hypothesis and the alternative hypothesis are two opposing statements used in hypothesis testing to make conclusions about a population parameter.

1. Null Hypothesis :

It is the statement that there is no effect, no difference, or no change.

It represents the status quo or the assumption that any observed difference is due to chance.

It is the hypothesis that is tested and possibly rejected.

2. Alternative Hypothesis :

It is the statement that contradicts the null hypothesis.

It suggests that there is an effect, a difference, or a change.

***`Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.`***

***`Answer-`***

The significance level is the maximum probability of making a wrong decision by rejecting a true null hypothesis. It helps decide whether the evidence from sample data is strong enough to support the alternative hypothesis.

***`Question 4: What are Type I and Type II errors? Give examples of each.`***

***`Answer-`***

1. Type I Error (False Positive):

Occurs when the null hypothesis is true, but we wrongly reject it.

It means we conclude there is an effect or difference when actually there isn’t one.

The probability of making a Type I error is denoted by (alpha) — the significance level.

Example:
Suppose a company claims a medicine has no side effects.

H₀: The medicine has no side effects.
H₁: The medicine has side effects.
If we conclude that the medicine has side effects (reject H₀) when in reality it does not, that’s a Type I error.

2. Type II Error (False Negative):

Occurs when the null hypothesis is false, but we fail to reject it.

It means we conclude there is no effect when actually there is one.

The probability of making a Type II error is denoted by  (beta).

Example:
Using the same medicine case:

H₀: The medicine has no side effects.

H₁: The medicine has side effects.
If we conclude that the medicine has no side effects (fail to reject H₀) when it actually does, that’s a Type II error.


***`Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.`***

***`Answer-`***

Both Z-test and T-test are statistical tests used to determine whether there is a significant difference between sample data and population parameters, or between two sample means.
However, they differ mainly in the sample size and whether the population standard deviation is known.

1. Z-Test:

Definition:
A Z-test is used when the population variance or standard deviation is known, and the sample size is large (n ≥ 30).

2. T-Test:

Definition:
A T-test is used when the population standard deviation is unknown, and the sample size is small (n < 30).

```python
'''
Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
(Include your Python code and output in the code box below.)
'''


import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Known population mean and standard deviation
mu_0 = 50      # Claimed population mean
sigma = 0.5    # Population standard deviation

# Sample statistics
x_bar = np.mean(sample_data)
n = len(sample_data)

# Calculating Z-statistic
z_stat = (x_bar - mu_0) / (sigma / np.sqrt(n))

# Calculating p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Display results
print(f"Sample Mean = {x_bar:.3f}")
print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value = {p_value:.4f}")

# Significance level
alpha = 0.05

# Decision
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference.")
else:
    print("Fail to reject the null hypothesis. No significant difference found.")
```

```python
'''
Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)
'''


import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


np.random.seed(0)  # For reproducibility
mean = 100          # True mean
std_dev = 15        # True standard deviation
size = 200          # Sample size

data = np.random.normal(mean, std_dev, size)


sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
standard_error = sample_std / np.sqrt(size)


confidence = 0.95
z_critical = stats.norm.ppf(1 - (1 - confidence) / 2)  # 1.96 for 95%
margin_of_error = z_critical * standard_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error


print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")


plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Normal Distribution (Simulated Data)")
plt.xlabel("Data values")
plt.ylabel("Frequency")


plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label='Sample Mean')
plt.axvline(ci_lower, color='green', linestyle='dotted', linewidth=2, label='95% CI Lower')
plt.axvline(ci_upper, color='green', linestyle='dotted', linewidth=2, label='95% CI Upper')
plt.legend()
plt.show()
```

```python
'''
Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.)
'''


import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores and plot histogram
def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculating Z-scores
    z_scores = (data - mean) / std_dev

    # Displaying first few Z-scores
    print("First 10 Z-scores:", np.round(z_scores[:10], 3))

    # Plotting histogram of Z-scores
    plt.hist(z_scores, bins=15, color='lightblue', edgecolor='black', alpha=0.7)
    plt.title("Histogram of Standardized Data (Z-scores)")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")


    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean (Z=0)')
    plt.axvline(1, color='green', linestyle='dotted', linewidth=2, label='Z=+1')
    plt.axvline(-1, color='green', linestyle='dotted', linewidth=2, label='Z=-1')
    plt.legend()
    plt.show()

    return z_scores

# Example dataset
data = np.random.normal(50, 10, 100)  # Mean=50, SD=10, 100 values

# Calling the function
z_scores = calculate_z_scores(data)
```
